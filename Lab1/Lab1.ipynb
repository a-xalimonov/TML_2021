## Текстовое описание набора данных

В качестве набра данных будем использовать встроенный в библиотеку scikit-learn набор данных 

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
%matplotlib inline 
sns.set(style="ticks")
```

```python
def make_dataframe(dataset_function):
    dataset = dataset_function()
    dataframe = pd.DataFrame(data= np.c_[dataset['data'], dataset['target']],
                     columns= list(dataset['feature_names']) + ['target'])
    return dataframe
```

```python
data = make_dataframe(load_boston)
```
